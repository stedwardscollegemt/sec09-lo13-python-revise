### 1. Dictionaries

A dictionary is another way you can store information in Python. It is similar to a list, but each item is stored as a `key:value` pair.

Dictionaries are written with curly brackets, and each item has two parts: a key and a corresponding value:

```python
dictionary = {
    'key1': 'value1',
    'key2': 'value2'
}
```
For example:

```python
favourite_foods = {
    'Simon': 'pizza',
    'Jill': 'pancakes',
    'Roger': 'custard'
}
```

### 2. Common Operations on Dictionaries

**Accessing and updating data in a dictionary**

Using dictionaries, we can easily access the data and process it. Similar to lists or tuple, we can access individual data items using the `[]` but this time, we specify the **key**!

Let us assume that Jill changed her mind — her favourite food is now pasta. Let us update her information:

```python
favourite_foods['Jill'] = 'pasta'
```

**Adding a new item to the dictionary**

We can use the same principle to add to the dictionary. We just need to make sure the **key** is unique. Let us add Julie to our dictionary. She likes to eat hamburgers.

```python
favourite_foods['Julie'] = 'hamburgers'
```

**Important built-in functions**

Since dictionaries are editable, Python offers various built-in functions to make processing them easier. Here is a list of functions you need to know about:

- `clear()` - removes all the elements from a dictionary
- `items()` - returns something to see all key-value pairs of the dictionary
- `get(key)` - returns the value of the item with the specified key
- `pop(key)` - removes the specified item from the dictionary

````python
# Example of a dictionary and simple operations
flag_colours = {
    'USA': ['blue', 'red', 'white'],
    'Austria': ['red', 'white'],
    'Italy': ['green', 'white', 'red'],
    'Germany': ['black', 'red', 'yellow'],
    'Spain': ['red', 'yellow'],
    'India': ['orange', 'white', 'green']
}

print(flag_colours.items())
print(flag_colours.get('Spain'))
````

### 3. Filtering Data Using Logical Operators: AND, OR, NOT

Using dictionaries to organise our data makes it convenient to write simple programs that can filter the data. 

Much like lists, we can use `for` loops to access each key/value pair in a dictionary using the `in` operator and the `items` function we learned about earlier.

**Looping through a dictionary**
```python
for country, colours in flag_colours.items():
    print(country, colours)
```

**Using other logical operators to filter data**

Logical operators are used to combine conditional statements. Python has three logical operators:

`and` - Returns `True` if both statements are true
`or` - Returns `True` if one of the statements is true
`not` - Reverses the result, returns `False` if the result is true

These function like our computer **logic gates**.

For example, we would like to find out which countries have the colours red **or** white in their flag.

```python
for country, colours in flag_colours.items():
    if 'red' in colours or 'white' in colours:
        print(country)
```
