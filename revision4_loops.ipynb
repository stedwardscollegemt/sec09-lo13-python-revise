#### 1. Introduction

Loops allow a program to repeat actions multiple times. Python provides two main types of loops:

`for` loops – used to iterate over a sequence (like a list or range).

`while` loops – used when a condition must remain `True`.

#### 2. Code Examples

**Example 1 - Simple for loop**

```python
for counter in range(5):
    print("Number of times:", i)
```

**Example 2 - For loop with a list**

```python
fruits = ["apple", "banana", "cherry"]
for fruit in fruits:
    print("I like", fruit)
```

**Example 3 - While loop**

```python
count = 0
while count < 3:
    print("Count is:", count)
    count += 1
```

**Example 4 - Loops with user input**

```python
# For loop
num = int(input("Enter a number: "))
for i in range(1, num + 1):
    print(i)

# While loop
response = ""
while(response != "Exit"):
    response = input("What would you like to do? Play, Study, or, Exit")
    if (response == "Exit"):
        print("Ok, bye.")
    else:
        print("Ok, let's", response)
```

**Example 5 - Infinite loop with break**

```python
while True:
    num = int(input("Enter option 1 - 4 or -1 to exit: "))
    if (num == -1):
        break # get out of the loop
    print(f"You chose option: {num}.")
```

#### 3. Nested Loops

Nested loops mean having one loop inside another.

**Example 6: Nested For Loop**

```python
for i in range(1, 4):
    for j in range(1, 3):
        print("i =", i, ", j =", j)
```

**Example 7: Nested While Loop**

```python
i = 1
while i <= 3:
    j = 1
    while j <= 2:
        print("i =", i, ", j =", j)
        j += 1
    i += 1
```
